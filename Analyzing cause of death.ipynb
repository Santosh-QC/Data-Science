```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
causes = ['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others']
percentile = [62, 5, 4, 2, 1 , 26]
plt.figure(figsize=(10,10))
explode = [0.5,0,0,0,0,0]
```

```python
plt.pie(percentile, explode=explode, labels=causes, autopct='%1.1f%%', startangle=70) 
plt.axis('equal')
plt.title('ohio state death')
```
