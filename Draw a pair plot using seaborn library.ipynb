```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
```

```python
auto_data = pd.read_csv('auto_data.csv')
auto_data.head()
```

```python
#user defined function for origin

def origin(num):
    if num==1:
        return 'USA'
    if num==2:
        return 'eurpoe'
    if num==3:
        return 'Asia'
    
auto_data['origin'] = auto_data['origin'].apply(origin)
        
```

```python
auto_data.head()
```

```python
sns.pairplot(auto_data[['mpg', 'weight', 'origin']], hue='origin', size=10)
```
