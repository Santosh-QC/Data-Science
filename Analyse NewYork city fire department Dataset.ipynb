```python
import pandas as pd

df_fire_dept = pd.read_csv('FDNY.csv', skiprows=1)
df_fire_dept.head()
```

```python
#total number of facilities

df_total_facility = df_fire_dept.FacilityName.count()
df_total_facility
```

```python
# number of fire departments in each of borough

df_fire_dept_borough = df_fire_dept.Borough.value_counts()
df_fire_dept_borough
```

```python
#facility name in manhattan

facility_manhattan = df_fire_dept.groupby('Borough')
```

```python
all_manhattan_facility = facility_manhattan.get_group('Manhattan')
all_manhattan_facility
```
